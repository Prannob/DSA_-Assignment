# Step 3) Data preprocessing

We can drop all feature which are irrelevant to model like  loan_id ,insurance_type,origination_date (data of loan started),
To understand this dropping we need to understand each feature in the data set so that we can understand relevant and irrelevant feature. 

We can also assign X (feature variable) and y(target variable) At this point also.
y is the target group here "m13" and all other features after dropping  unnecessary features and target varibale can be our X 

```python
data.columns # TO get the features of the dataset
```

```python
X= data.drop(['loan_id','origination_date','first_payment_date','insurance_percent','insurance_type', 'm13'],axis=1)
# To drop column wise data give axis =1
y= pd.DataFrame(data["m13"])
```

One hot code encoding is done to "Source" and "loan_purpose"

```python

print(X["source"].value_counts())
print ("\n")
print(X["loan_purpose"].value_counts())
X
```

```python
# One Hot Encoding in "source" and  "loan_purpose"
source_en= pd.get_dummies(X["source"])
loanp_en= pd.get_dummies(X["loan_purpose"])
```

So now we have encoded data of "source" and "loan_purpose" in two dataframe source_en and loanp_en respectively.
We need to add this encoded dataframe with X and drop the original "source" and "loan_purpose" from X

```python
X.info()
```

```python
X=X.drop(["source","loan_purpose"],axis=1) #drop the original "source" and "loan_purpose" from X
```

```python
X
```

```python
X=pd.concat([X,source_en,loanp_en],axis=1) # Concat/ add encoded dataframe with X
#dataframe source_en and loanp_en respectively.We need to add this encoded dataframe with X 
```

```python
# Label encording is done in "financial_institution".
```

```python
# Label Encoding of feature "financial_institution"
from sklearn.preprocessing import LabelEncoder
label_en= LabelEncoder()
a= ['financial_institution'] # 1st Label encoding is done in 'financial_institution'
for i in np.arange(len(a)):
    X[a[i]]= label_en.fit_transform(X[a[i]])
```

# Step 4 ) Feature Engineering

Feature Engineering means new features are generated from the exsisting features. Data is not imported from outside. 
Simply Generating new data from exisiting features but adding two features or substracting or geting ratio etc

After creating new features , if needed those features which is used to create the new features can be dropped . Thus reduce our fetures but still have all the relevant data  


```python
X.columns
```

```python
X["credit score"]= X['co-borrower_credit_score']+X['borrower_credit_score'] 
# New feature "credit score" is created from  'borrower_credit_score' and 'co-borrower_credit_score'
X=X.drop([ 'borrower_credit_score','co-borrower_credit_score'],axis =1)
# The  'borrower_credit_score' and 'co-borrower_credit_score' is dropped from X but the effect of the feature still present in X
```

```python
X.columns
```

We can see new Column  'credit score' is found in the X but not 'borrower_credit_score' and 'co-borrower_credit_score'

```python
# SIMILARLY WE CAN DO OTHER COLUMNS AS WELL
```

```python
X["mean"]= X[['m1', 'm2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8','m9', 'm10', 'm11', 'm12']].mean(axis=1)
X["sum"]= X[['m1', 'm2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8','m9', 'm10', 'm11', 'm12']].sum(axis=1)
X["skew"]= X[['m1', 'm2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8','m9', 'm10', 'm11', 'm12']].skew(axis=1)
X["kurt"]= X[['m1', 'm2', 'm3', 'm4', 'm5', 'm6', 'm7', 'm8','m9', 'm10', 'm11', 'm12']].kurt(axis=1)
```

# Step 5) Building the model

Before building the model we need to split the data for training and testing

```python
# Splitting the data set to training set and testing set

from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test= train_test_split(X,y,random_state=2,test_size=.3)
# test_size=.3 means  30% of data is taken for testing and remaining 70 % data is taken for training 
# Testing set should be les than training usually taken as  30  or below percentage
```

```python
print("dimention of X_Test is :", X_test.shape)
print("dimention of X_Train is :", X_train.shape)
print("dimention of y_Test is :", y_test.shape)
print("dimention of y_Train is :", y_train.shape)
```

```python
# Building model (We can use any model creation but here we are using Random Forest)
from sklearn.ensemble import RandomForestClassifier # Import the library
rf= RandomForestClassifier() # Create the instance of the data
rf.fit(X_train,y_train) # Training the model
y_pred= rf.predict(X_test) # Making the prediction 
```

Now we have created the model 

# step 6) To evaluate the the performance of the model 

```python
from sklearn.metrics import confusion_matrix, precision_score, accuracy_score, recall_score,f1_score
```

```python
print("Accuracy is ",accuracy_score(y_test, y_pred))
print("Precision is ",precision_score(y_test, y_pred))
print("Recall is ",recall_score(y_test, y_pred))
print("F1 score is ",f1_score(y_test, y_pred))
```

Value of f1 score varies between 0 and 1 . If the value is close to 1 then model is very good, if it is near to zero then it is not a good model

```python
confusion_matrix(y_test, y_pred)
```

Here 128 and 27 is the False positive value and False negative value in our model

# Step 8) Distribution of Prediction Probability

```python
threshold= 0.22 # default threshold value is 0.5 but here it is changed to .22
y_pred_proba= rf.predict_proba(X_test)[:,1] # Making the prediction 
# instead of rf.predict in building model session here we use "rf.predict_proba"  
# we will be calculating the probabilites of 0s and 1s
y_pred=(y_pred_proba>threshold).astype(int) # predicting the value with condition
# compare the probabilty value with threshold value
# eg if probablity values is 0.55 with threshold value 0.22 if the condition is satsified it convert it into int ie if condition
# is true then we get value as 1 .
```

```python
print("F1 score is ",f1_score(y_test, y_pred))
```

we check if this affects the performance of model by cheching th f1 score . We can see that f1 score increased from 0.48 (previous step) to .53. That is perfomance of the model or efficency of the model  is increased
