```python
#call numpy, pandas, matplotlib and seaborn fucntion
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Q1 ) #to import csv file saved in the local Disk to python environment.
car= pd.read_csv("cars_data.csv")
car
```

```python
# Q2) To check for the null values present
# At first info of the DataFrame is checked
car.info()
```

```python
# It is found that there are no null value present in the data
# Or sum of null valus can be add using the below equation
# If any column have null value it give a value greater than zero.
car.isna().sum()
```

```python
#Q3) Plot a bar graph of male vs female buyers participated in the sales
plt.figure()
plt.hist(car['Buyer Gender'])
plt.title(" Bar graph of male vs female buyers participated in the sales")
plt.xticks()
plt.yticks()
plt.ylim(4900,5200)
plt.tight_layout()
"\n"
print("Number of Female and male Participants")
car['Buyer Gender'].value_counts()
```
